Following code will make pandas Dataframe out of web table:

```python
import pandas as pd
import lxml
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
dfs = pd.read_html(url)[0]
df=pd.DataFrame(dfs)
df.head()
```

Following code will remove cells with non assigned Borough.

```python
df.drop(df.loc[df['Borough']=='Not assigned'].index, inplace=True)
df.head()
```
